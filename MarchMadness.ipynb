```python
# Read the data from the file
with open('SeedsVSRounds.csv', 'r') as file:
    lines = file.readlines()

# Group the games data by seed
seed_data = {}
for line in lines:
    seed, game = line.strip().split(',')
    if seed != 'Seed' and game != 'Games':
        if seed not in seed_data:
            seed_data[seed] = []
        seed_data[seed].append(int(game))


def hist(seed):
    distribution = seed_data[str(seed)]
    plt.hist(distribution, bins=range(1, 9), rwidth=.8)
    
    plt.xlabel('Round Score')
    plt.ylabel('Frequency')
    plt.title(f'{seed} Seed')
    
    plt.show()

hist(1)
hist(2)
hist(3)
hist(4)
hist(5)
hist(6)
hist(7)
hist(8)
hist(9)
hist(10)
hist(11)
hist(12)
hist(13)
hist(14)
hist(15)
hist(16)
```

```python
# Read the data from the file
with open('SeedsVSRounds.csv', 'r') as file:
    lines = file.readlines()
    
# Group the games data by seed
seed_data = {}
for line in lines:
    seed, game = line.strip().split(',')
    if seed != 'Seed' and game != 'Games':
        if seed not in seed_data:
            seed_data[seed] = []
        seed_data[seed].append(int(game))

# Calculate the mean number of games for each seed
seed_means = {}
for seed, games in seed_data.items():
    seed_means[seed] = np.mean(games)


    
def avg(x):
    return 1 + 5/x + 3/x**2 - 4/x**3 + 3/x**4

    
x_values = []
y_values = []    
# Print the results
for seed, mean in seed_means.items():
    x_values.append(int(seed))
    y_values.append(float(mean))
    
plt.scatter(x_values, y_values)

# Add labels and title
plt.xlabel('Seed')
plt.ylabel('Games')
plt.title('Scatter Plot of Seed vs Games')

x_range = np.linspace(1, 16, 1500)
plt.plot(x_range, avg(x_range), 'r-')

# Display the plot
plt.show()
```
